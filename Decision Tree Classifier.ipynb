```python
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn import tree

iris = datasets.load_iris()
iris
```

```python
type(iris)
```

```python
iris.keys()
```

```python
X = iris['data']
X
```

```python
y = iris['target']
y
```

```python
y_classes = iris['target_names']
y_classes
```

```python
column_names = iris['feature_names']
column_names
```

```python
df = pd.DataFrame(X, columns = column_names)
df
```

```python
df['target'] = y
df
```

```python
df['target'] = df['target'].apply(lambda x: y_classes[x])
df
```

```python
X = df.iloc[:, :-1]
X
```

```python
X = df.drop('target', axis = 1)
X
```

```python
y = df['target']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = tree.DecisionTreeClassifier(criterion = 'gini', random_state = 0)
model
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

train_acc, test_acc
```

```python
# plotting the decision tree in a jpg
from six import StringIO
import pydot
tree.export_graphviz(model, out_file = 'tree.dot')
out_data = StringIO()
tree.export_graphviz(model, out_file = out_data,
                     feature_names = column_names,
                     class_names = model.classes_,
                     filled = True, rounded = True,
                     special_characters = True,
                     node_ids = 1)
graph = pydot.graph_from_dot_data(out_data.getvalue())
graph[0].write_jpg('iris.jpg')
```
